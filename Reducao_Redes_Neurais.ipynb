```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
# Carregar imagem colorida
def load_image(path):
    img = Image.open(path).convert("RGB")  # Garante que a imagem esteja no modo RGB
    return img
```

```python
# Converter para tons de cinza
def rgb_to_grayscale(img):
    return img.convert("L")  # Converte diretamente para escala de cinza
```

```python
# Binarizar imagem
def grayscale_to_binary(img, threshold=128):
    img_array = np.array(img)
    binary_array = (img_array > threshold) * 255  # Mantém os valores como 0 ou 255
    return Image.fromarray(binary_array.astype(np.uint8))
```

```python
# Carregar imagem colorida do coelho executivo
image_path = '/content/executive rabbit.png'  # Ajuste o caminho conforme necessário
color_image = load_image(image_path)
```

```python
# Mostrar imagem colorida
plt.imshow(color_image)
plt.axis("off")
plt.title("Imagem Colorida")
plt.show()
```

```python
# Converter para tons de cinza
gray_image = rgb_to_grayscale(color_image)
```

```python
# Mostrar imagem em tons de cinza
plt.imshow(gray_image, cmap='gray')
plt.axis("off")
plt.title("Imagem em Tons de Cinza")
plt.show()
```

```python
# Binarizar imagem
binary_image = grayscale_to_binary(gray_image)
```

```python
# Mostrar imagem binarizada
plt.imshow(binary_image, cmap='gray')
plt.axis("off")
plt.title("Imagem Binarizada")
plt.show()
```
